```python
#necessary dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
#mounting google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#importing the dataset
dataset = pd.read_csv(r'/content/drive/MyDrive/letter-recognition.csv')
dataset.head()
```

```python
#declaring x and y variables
x_vars = dataset.iloc[:,1:]
print(x_vars)
y_vars = dataset['letter']
y_vars = y_vars.replace({'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7, 'H':8,'I':9, 'J':10,'K':11,'L':12,'M':13,'N':14,'O':15,'P':16,'Q':17,'R':18,'S':19,'T':20,'U':21, 'V':22, 'W':23,'X':24,'Y':25,'Z':26}).astype(float)
```

```python
#testing and training splut
x_train,x_test,y_train,y_test = train_test_split(x_vars,y_vars,train_size = 0.7,random_state=42)
```

```python
print(type(y_train))
```

```python
#fitting the SVM model and printing confusion matrix for training data
svm_fit = SVC(kernel='linear',C=1.0,random_state=43)
svm_fit.fit(x_train,y_train)
print ("\nSVM Linear Classifier - Train Confusion Matrix\n\n",pd.crosstab(y_train, svm_fit.predict(x_train),rownames =["Actuall"],colnames = ["Predicted"]) )
print ("\nSVM Linear Classifier - Train accuracy:",round(accuracy_score(y_train, svm_fit.predict(x_train))*100,3))
print ("\nSVM Linear Classifier - Train Classification Report\n", classification_report(y_train,svm_fit.predict(x_train)))
```

```python
#fitting for testing dataset and printiong the confusion matrix for the same
svm_fit = SVC(kernel='linear',C=1.0,random_state=43)
svm_fit.fit(x_test,y_test)
print ("\nSVM Linear Classifier - Train Confusion Matrix\n\n",pd.crosstab(y_test, svm_fit.predict(x_test),rownames =["Actuall"],colnames = ["Predicted"]) )
print ("\nSVM Linear Classifier - Train accuracy:",round(accuracy_score(y_test, svm_fit.predict(x_test))*100,3))
print ("\nSVM Linear Classifier - Train Classification Report\n", classification_report(y_test,svm_fit.predict(x_test)))
```

```python
#using polynomial kernal to train
svm_poly_fit = SVC(kernel='poly',C=1.0,degree=2)
svm_poly_fit.fit(x_train,y_train)
print ("\nSVM polynomial kernel Classifier - Train Confusion Matrix\n\n",pd.crosstab(y_train, svm_poly_fit.predict(x_train),rownames =["Actuall"],colnames = ["Predicted"]) )
print ("\nSVM  polynomial kernel Classifier - Train accuracy:",round(accuracy_score(y_train, svm_poly_fit.predict(x_train))*100,3))
print ("\nSVM  polynomial kernel Classifier - Train Classification Report\n", classification_report(y_train,svm_poly_fit.predict(x_train)))
```

```python
#using polynomial kernal to test
print ("\n\nSVM Polynomial Kernel Classifier - Test Confusion Matrix\n\n", pd.crosstab(y_test,svm_poly_fit.predict(x_test),rownames = ["Actuall"],colnames = ["Predicted"]))
print ("\nSVM Polynomial Kernel Classifier - Test accuracy:",round(accuracy_score( y_test,svm_poly_fit.predict(x_test))*100,3))
print ("\nSVM Polynomial Kernel Classifier - Test Classification Report\n", classification_report(y_test,svm_poly_fit.predict(x_test)))
```

```python
#using RBF kernal to train
svm_rbf_fit = SVC(kernel='rbf',C=1.0, gamma=0.1)
svm_rbf_fit.fit(x_train,y_train)
print ("\nSVM RBF Kernel Classifier - Train Confusion Matrix\n\n",pd.crosstab( y_train,svm_rbf_fit.predict(x_train),rownames = ["Actuall"],colnames = ["Predicted"]))
print ("\nSVM RBF Kernel Classifier - Train accuracy:",round(accuracy_score( y_train, svm_rbf_fit.predict(x_train))*100,3))
print ("\nSVM RBF Kernel Classifier - Train Classification Report\n", classification_report(y_train,svm_rbf_fit.predict(x_train)))
```

```python
#using RBF kernal to test
print ("\n\nSVM RBF Kernel Classifier - Test Confusion Matrix\n\n", pd.crosstab(y_test,svm_rbf_fit.predict(x_test),rownames = ["Actuall"],colnames = ["Predicted"]))
print ("\nSVM RBF Kernel Classifier - Test accuracy:",round(accuracy_score( y_test,svm_rbf_fit.predict(x_test))*100,3))
print ("\nSVM RBF Kernel Classifier - Test Classification Report\n", classification_report(y_test,svm_rbf_fit.predict(x_test)))
```
